# Iris identification

https://archive.ics.uci.edu/dataset/53/iris

| Variable     | Role        | Type       | Description | Units |
| - | - | - | - | - |
| Sepal length | Feature | Continuous | | cm |
| Sepal width  | Feature | Continuous | | cm |
| Petal length | Feature | Continuous | | cm |
| Petal width  | Feature | Continuous | | cm |
| Class | Target | Categorical | Class of plant: Iris Setosa, Iris Versicolour, or Iris Virginica | |

```python
import numpy as np
np.set_printoptions(legacy='1.25')
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
col_names = ['Sepal_length', 'Sepal_width', 'Petal_length', 'Petal_width', 'Iris_class']
iris_data = pd.read_csv('iris/iris.data', header=None, names=col_names, sep=",")
iris_data
```

```python
# checking the occurrence of each class - they should be approximately equal
iris_data['Iris_class'].value_counts().plot(kind='bar')
```

## Feature pair plot

This can give us a good idea of whether certain features will be particularly useful for classifying.

```python
sn.pairplot(iris_data, hue ='Iris_class')
```

```python
X = iris_data[col_names[0:4]]
y = iris_data[col_names[4]]

# split data into training and test sets
# sklearn default is 75% training, 25% test
X_train, X_test, y_train, y_test = train_test_split(X.values, y, random_state = 0)

# create the KNN classifier:
knn = KNeighborsClassifier(n_neighbors = 1)

# pass training set to classifier
knn.fit(X_train, y_train)

print('Accuracy of k-NN classifier: ', knn.score(X_test, y_test))
```

```python
iris_data.Iris_class.unique()

knn.predict([[5.5, 2.2, 10, 0.70]])[0]
```
